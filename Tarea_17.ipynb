## Ejercicio 1: ¿Como podemos evaluar funciones en matrices?
__Hint:__ ¿que operaciones sabemos que estan bien definidas sobre matrices?

Sean $A, B, C ∈ M_{n×m}$ 

Las operaciones que están bien definidas sobre matrices son las que se mencionan a continuación:

### Suma de matrices

La suma de matrices cumple con las siguientes propiedades:

Cerradura:

La suma de dos matrices de orden $m x n$ es otra matriz con la misma dimensión que $A$ y $B$, es decir, $m x n$.

Conmutatividad:

$$A + B = B + A$$

Asociatividad:

$$A + (B + C) = (A + B) + C$$

Elemento neutro:

$$A + 0 = A$$

Inverso aditivo:

$$A + (−A) = 0$$

Donde 0 es la matriz nula, cuya dimensión es la misma que la de la matriz A.

### Producto de un escalar por una matriz

Considerando una matriz $A=(a_{ij})$ y un número real K∈R, se define el producto de un número real por una matriz como la matriz del mismo orden que A, cuyas entradas están multiplicadas por K.

$$KA=(Ka_{ij})$$

Propiedades del producto de un escalar por una matriz con $ A ∈ Mmxn, a, b ∈ \Re $

Asociatividad de la multiplicación:

$$a ·  (b · A) = (a · b) · A $$ 

Distributividad:

$$a  ·  (A + B) = a · A + a · B $$ 

$$(a + b) · A = a · A + b · A $$  

Identidad multiplicativa:

$$1 · A = A$$

### Producto de matrices

Para que dos matrices se puedan multiplicar, se tiene que cumplir que el número de columnas de A sea igual que el número de filas de B.

$$M_{m×n} × M_{n×n}=M_{m×p}$$ 

El elemento $c_{ij}$ de la matriz producto se obtiene multiplicando cada elemento de la fila i de la matriz A por cada elemento de la columna j de la matriz B y sumándolos.

Propiedades del producto de matrices

Asociatividad:

$$A · (B · C) = (A · B) · C$$

Distributividad del producto respecto de la suma:

$$A · (B + C) = A · B + A · C$$

Elemento neutro:

$$A · I = A$$

Donde I es la matriz identidad que tiene el mismo orden que la matriz A.

La multiplicación de matrices no es conmutativa:

$$A · B ≠ B · A$$

## Ejercicio 2: ¿Como se define una matriz simétrica?

Una matriz A cuadrada es simétrica si y solo si A es igual a su traspuesta, es decir, 

$$ A={A}^{T}$$

## Ejercicio 3: ¿Como puede construir una matriz simétrica a partir de cualquier matriz cuadrada?

No toda matriz cuadrada es una matriz simétrica, pero podemos intentar diagonalizar una matriz cuadrada para obtener una matriz simétrica. 

Un teorema establece que una matriz $A$, de orden n, es diagonalizable si y sólo si tiene un conjunto $\overrightarrow { { x }_{ 1 } }, \overrightarrow { { x }_{ 2 } },...,\overrightarrow { { x }_{ n } } $de n vectores linealmente independientes, de donde:

$${ P }^{ -1 }AP=\left( \begin{matrix} { \lambda  }_{ 1 } & 0\quad \quad ... & 0 \\ 0 & { \lambda  }_{ 2\quad \quad  }... & 0 \\ ... & ...\quad ... & ... \\ 0 & 0\quad \quad ... & { \lambda  }_{ n } \end{matrix} \right) $$

$sii$ $$A=P\left( \begin{matrix} { \lambda  }_{ 1 } & 0\quad \quad ... & 0 \\ 0 & { \lambda  }_{ 2\quad \quad  }... & 0 \\ ... & ...\quad ... & ... \\ 0 & 0\quad \quad ... & { \lambda  }_{ n } \end{matrix} \right){ P }^{ -1 }$$


Una matriz cuadrada $A$, de orden n, se dice que es diagonalizable si existe una matriz invertible P, de orden n, y una matriz diagonal $D$, tal que $P^{-1} AP = D$, lo cual es equivalente a decir que una matriz cuadrada A es diagonalizable si es semejante a una
matriz diagonal.

Lo que se busca al  diagonalizar una matriz, es encontrar las matrices $D$ y $P$ tales que:

$$ A= PDP^{-1}  $$

las columnas de la matriz P son los vectores $\overrightarrow { { x }_{ 1 } }, \overrightarrow { { x }_{ 2 } },...,\overrightarrow { { x }_{ n } } $ y $\overrightarrow { { \lambda }_{ 1 } }, \overrightarrow { { \lambda  }_{ 2 } },...,\overrightarrow { { \lambda  }_{ n } } $ son los valores propios de A.

## Ejercicio 4: Construya el propagador $e^{A t}$ para cada tiempo sobre una malla y uselo para propagar un estado inicial $\vec x_0$ cualquiera. Utilice:
$$A=\left(\begin{array}{cc} -1 & 1\\ -1 & -1 \end{array} \right).$$
* Grafique la trayectoria en el espacio de fase.
* Recuerde que $\vec x(t) = e^{A t}\vec x_0$, y necesitamos graficar la trayectoria $\vec x(t)$ en el espacio de fase.

El propagador $e^{A t}$ está dado por:

```julia
A=[-1 1; -1 -1]     # Definimos nuestra matriz A
```

```julia
vals = eigvals(A);  # Valores propios de la matriz A.
w = eigvecs(A);     # Matriz cuyas columnas son los vectores propios.
d = diagm(vals);    # Matriz cuyas columnas son los valores propios.
```

```julia
eA(t) = w*diagm(exp.(vals*t))*inv(w)       # El propagador para cada tiempo sobre una malla
```

```julia
E(t,x0) = w*diagm(exp.(vals*t))*inv(w)*x0  # Multiplicando por la condición inicial
```

```julia
using PyPlot
t1 = linspace(0,15,100)             # Creamos un intervalo con los tiempos a considerar
for i in 1:500                      # Ciclo for para graficar 500 condiciones iniciales
    X = []                          # Arreglo vacío donde se guardará el propagador para cada tiempo, utilizando una determinada condición.
    x0 = [2*rand()-1; 2*rand()-1]   # Tomando condiciones al azar
 
    for k in 1:length(t1)           # Ciclo for que va a recorrer la longitud del intervalo de tiempo
        e = E(t1[k], x0)            # Evaluando el propagador en el tiempo k
        push!(X,e)                  # Arregando el resultado al arreglo vacío, creado inicialmente
    end

    listx = map(x->x[1],real(X))    # Creamos unas listas para x y u
    listu = map(x->x[2],real(X))
    plot(listx,listu)               # Graficando las listas
end
title("Espacio fase")
xlabel("x(t)")
ylabel("u(t)")

show()
```
